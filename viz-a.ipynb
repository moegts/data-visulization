# **Visulization** **Plots** about **Flights**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# plt.plot()
def plot (y1, y2):
    plt.figure()
    plt.plot(y1, label = 'data from y1')
    plt.plot(y2, label = 'data from y2')
    plt.legend()
    plt.title('Seaborn')
    plt.xlabel('Defult index')
    plt.ylabel('Data')
    plt.show()
plot([5,6,9,3,-5,10],[10,2,5,9,5,1])   
```

```python
sns.set()
sns.get_dataset_names()
```

```python
data = sns.load_dataset('flights')
data.head(10)
```

```python
sns.relplot(x='year', y='passengers', data=data, kind='line',aspect=2.5)
```

```python
sns.set_theme(style="dark")
sns.relplot(x='year', y='passengers', hue='month', data = data)
sns.relplot(x='month', y='passengers', hue='year', data = data)
```

```python
groupedMonth = data.groupby('month').sum()
sns.relplot(x='month', y='passengers', data = groupedMonth)
```
