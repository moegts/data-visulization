# **Visulization** **Plots** about **Penguins**

```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
```

### set seaborn

```python
sns.set()
```

### Load the data

```python
df = sns.load_dataset('penguins')
df.head()
```

### Plot with **Linear Regrusion** `lmplot`

```python
sns.lmplot( x='bill_length_mm',
            y='body_mass_g',
            data=df)
```

### Plot as **PAR** first 15 record `barplot`

```python
sns.barplot(x='bill_length_mm',
            y='body_mass_g',
            data=df.head(15))
```

### Plot with Relational `relplot`

```python
sns.relplot(x='bill_length_mm',
            y='body_mass_g',
            data=df)
```

```python
sns.barplot(x='bill_length_mm',
            y='body_mass_g',
            data=df,
            palette="flare",
            ci=None)
```

```python
sns.relplot(x='bill_length_mm',
            y='body_mass_g',
            hue="sex", 
            data=df,
            palette="flare",)
```
